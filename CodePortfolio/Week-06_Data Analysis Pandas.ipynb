```python
#-- import pandas and necessary packages --- #

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#--- load the iris dataset from csv and assign headers(columns) -- #

df = pd.read_csv("iris.data",names=(["Sepal_length", "Sepal_width", "Petal_length", "Petal_width", "Class"  ]),header=None)
```

```python
# -- Check the first 5 rows ---#
df.head(5)
```

```python
#--- Check number of rows,columns using shape method ---#
df.shape
```

**Practice**

```python
# Dataframe returns rows where petal lenght is greater than 6
df.query('Petal_length > 6')
```

```python
# Dataframe returns rows where petal lenght is greater than 6 and Petal width > 2
df.query('Petal_length > 6 & Petal_width > 2') 
```

```python
# Dataframe returns rows where petal lenght is greater than 6 OR Petal Length > 1.5
df.query('Petal_length > 6 | Petal_length < 1.5')
```

```python
# Check if there are Null values in dataset.
df.isnull().sum()
```
